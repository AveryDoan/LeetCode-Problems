```python
#20 Valid parenthesis
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.height = 0

    def push(self, value):
        new_node = Node(value)
        if self.height == 0:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.height += 1

    def pop(self):
        if self.height == 0:
            return None
        else:
            temp = self.top
            self.top = self.top.next
            temp.next = None
            self.height -= 1
            return temp.value

    def is_empty(self):
        return self.height == 0

def is_valid_parenthesis(s):
    stack = Stack()
    for i in s:
        if i in ["(", "{", "["]:
            stack.push(i)
        elif i in [")", "}", "]"]:
            if stack.is_empty():
                return False
            top = stack.pop()
            if (i == ")" and top != "(") or \
               (i == "}" and top != "{") or \
               (i == "]" and top != "["):
                return False
        else:
            pass
    
    return stack.is_empty()


# def is_valid(s):
#     stack = Stack(10) 
#     stack.pop()
#     for p in s:
#         if p == "(":
#             stack.push(p)
#         elif p == ")":
#             if stack.is_empty() or stack.pop() != "(":
#                 return False
#     return stack.is_empty()


string = "()"
print(is_valid_parenthesis(string))  
```

```python
## 234 Palindrome
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class Stack:
    def __init__(self):
        self.top = None
        self.height = 0

    def push(self, value):
        new_node = Node(value)
        if self.height == 0:
            self.top = new_node
        else:
            new_node.next = self.top
            self.top = new_node
        self.height += 1

    def pop(self):
        if self.height == 0:
            return None
        else:
            temp = self.top
            self.top = self.top.next
            temp.next = None
            self.height -= 1
            return temp.value

    def is_empty(self):
        return self.height == 0

def Palindrome(s):
    stack = Stack()
    
    for char in s:
        stack.push(char)
    
    reversed_s = ""
    while not stack.is_empty():
        reversed_s += stack.pop()
    
    return s == reversed_s

string = "12214"
print(Palindrome(string))  
```

```python
# 938. Range Sum of BST

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None
class BinarySearchTree:
    def __init__(self):
        self.root = None
        
    def insert(self, value):
        new_node = Node(value) #initialize new node
        if self.root == None:
            self.root = new_node
            return True
        temp = self.root
        while True:
            if new_node.value == temp.value: #avoid duplicate
                return False 
            if new_node.value < temp.value:
                if temp.left == None:
                    temp.left = new_node
                    return True
                
                temp = temp.left

            if new_node.value > temp.value:
                if temp.right == None:
                    temp.right = new_node
                    return True
                
                temp = temp.right
                
    def range_sum(self, low, high):
        if self.root is None:
            return 0

        sum = 0
        queue = []
        queue.append(self.root)

        while queue:
            current_node = queue.pop(0)

            if current_node.value >= low and current_node.value <= high:
                sum += current_node.value

            if current_node.left is not None:
                if current_node.left.value >= low:
                    queue.append(current_node.left)
            if current_node.right is not None:
                if current_node.right.value <= high:
                    queue.append(current_node.right)

        return sum




my_tree = BinarySearchTree()
my_tree.insert(47)
my_tree.insert(21)
my_tree.insert(76)
my_tree.insert(18)
my_tree.insert(27)
my_tree.insert(52)
my_tree.insert(82)
print(my_tree.range_sum(20, 50))
```

```python
#617 MERGED BINARY
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

class BinarySearchTree:
    def __init__(self):
        self.root = None
        
    def insert(self, value):
        new_node = Node(value)
        if self.root is None:
            self.root = new_node
            return True
        temp = self.root
        while True:
            if new_node.value == temp.value:
                return False 
            if new_node.value < temp.value:
                if temp.left is None:
                    temp.left = new_node
                    return True
                temp = temp.left
            if new_node.value > temp.value:
                if temp.right is None:
                    temp.right = new_node
                    return True
                temp = temp.right
                
    def BFS(self):
        queue = []
        results = []
        current_node = self.root
        queue.append(current_node)

        while len(queue) > 0:
            current_node = queue.pop(0)
            results.append(current_node.value)
            if current_node.left is not None:
                queue.append(current_node.left)
            if current_node.right is not None:
                queue.append(current_node.right)

        return(results) 

    def merge_trees(self, tree1, tree2):
        if tree1 is None:
            return tree2
        if tree2 is None:
            return tree1

        merged_node = self.merge_nodes(tree1.root, tree2.root)
        merged_tree = BinarySearchTree()
        merged_tree.root = merged_node
        return merged_tree

    def merge_nodes(self, node1, node2):
        if node1 is None:
            return node2
        if node2 is None:
            return node1

        new_node = Node(node1.value + node2.value)
        new_node.left = self.merge_nodes(node1.left, node2.left)
        new_node.right = self.merge_nodes(node1.right, node2.right)
        return new_node



# Example usage:
my_tree1 = BinarySearchTree()
my_tree1.insert(47)
my_tree1.insert(21)
my_tree1.insert(76)

my_tree2 = BinarySearchTree()
my_tree2.insert(18)
my_tree2.insert(27)
my_tree2.insert(52)
my_tree2.insert(82)

merged_tree = BinarySearchTree()
merged_tree.root = merged_tree.merge_trees(my_tree1, my_tree2)
print(merged_tree.root.BFS())
```
